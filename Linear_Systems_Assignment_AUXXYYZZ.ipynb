```python
import numpy as np
import scipy.linalg as la
newTask(1.1)
I4 = np.eye(4,4)
print(I4)

newTask(1.2)
Q, R = la.qr(I4)
print("Q:\n", Q)
print("R:\n", R)
assert np.allclose(Q @ R, I4)
print(Q@R==I4)
```

```python
newTask(1.3)
# Q and R are not unique, so we can multiply them by -1 without changing their product
Q2 = -np.eye(4)
R2 = -np.eye(4)

print("Q2:\n", Q2)
print("R2:\n", R2)
print(Q2@R2 == I4) 
```

```python
newTask(1.4)
Z4 = np.zeros((4,4))
Q3, R3 = la.qr(Z4)
print("Q3:\n", Q3)
print("R3:\n", R3)
print(Q3@R3 == Z4)
# Since Z4 is the zero matrix, the QR factorization of Z4 is not unique. One possible factorization is Q3 = I4 and R3 = Z4, which satisfies Q3 @ R3 = Z4. Another possible factorization is Q3 = Z4 and R3 = I4, 
# which also satisfies Q3 @ R3 = Z4. 
# Therefore, we have multiple valid QR factorizations for the zero matrix. In fact, any orthogonal matrix Q3 and any upper triangular matrix R3 such that Q3 @ R3 = Z4 would be a valid QR factorization of Z4.
Q4 = np.zeros((4,4))
R4 = np.array([[5000, 0, 0, 0], [0, 5000, 0, 0], [0, 0, 5000, 0], [0, 0, 0, 5000]])
print("Q4:\n", Q4)
print("R4:\n", R4)
print(Q4@R4 == Z4)
```

# Research: Least Squares on Trump Tweets from 2020

Back in 2.g at HTX, I made some "Trump"-tweet-analysis with my group on whether the number of likes could correlate with specific words.
Our conclusion back then was pretty simple based on the data and fundamental math we used, that there was certainly certain words that resulted in more likes on average. However we used a word-average algorithm, so I have found the data I scraped back then and will analyze whether the following had something to do with the amount of Retweets and Likes a trump tweet got:

* Date / time for whether it was best for engagement to tweet in the afternoon vs. morning

* Text length 

* Hashtags (boolean)

This time, I will construct a design matrix and apply Least Squares regression to quantify the relationship between these features and engagement.
## Sources:

Original HTX project: https://github.com/Jdreioe/TrumpTweet/
  
Lstsq: https://numpy.org/doc/stable/reference/generated/numpy.linalg.lstsq.html

Pandas: https://pandas.pydata.org/docs/


I'll start by loading the dataset into a pandas dataframe and filter empty tweets (eg. just photos/memes)

```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
file_path = "trump_tweets.csv"
trump_latest = pd.read_csv(file_path, sep=',', encoding='utf8', quotechar='"')
# Removing rows where the 'Text' column is empty or contains only whitespace to ensure that we only analyze tweets with actual content.
trump_latest = trump_latest[trump_latest['Text'].fillna('').str.strip() != '']

text = trump_latest['Text'].values.tolist()

# Transforming Likes and Retweets using log1p to handle zero values and reduce skewness in the data, which is common in social media metrics.
Likes = np.log1p(trump_latest['Likes'].values.tolist())
Retweets = np.log1p(trump_latest['Retweets'].values.tolist())
date = trump_latest['Formatteret_Date'].values.tolist()
hashtags = (trump_latest['Hashtags'].fillna('') != '').values.tolist()
text_length = [len(t) for t in text]

hour_of_day = [datetime.strptime(d, "%a %b %d %I:%M:%S %p +0000 %Y").hour for d in date]
```

```python
A = np.column_stack([
    np.ones(len(text)),        # bias (konstantled)
    text_length,               # tweet-længde
    hour_of_day,               # time på dagen (0-23)
    hashtags                   # boolean (True/False → 1/0)

])
b_likes    = np.array(Likes,    dtype=float)
b_retweets = np.array(Retweets, dtype=float)

# Least Squares solution
x_likes,    _, _, _ = np.linalg.lstsq(A, b_likes,    rcond=None)
x_retweets, _, _, _ = np.linalg.lstsq(A, b_retweets, rcond=None)

print("Coefficients for Likes:")
print(f"  Bias:        {x_likes[0]:.2f}")
print(f"  Text length: {x_likes[1]:.2f}")
print(f"  Hour of day: {x_likes[2]:.2f}")
print(f"  Hashtag:     {x_likes[3]:.2f}")

import matplotlib.pyplot as plt

# Predicted values based on the Least Squares solution @ with the feature matrix A
pred_likes    = A @ x_likes
pred_retweets = A @ x_retweets

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle("Trump Tweet Engagement Analysis pt. 1", fontsize=16)

# ── LIKES ────────────────────────────────────────────────────────────────────

# Predicted vs. Actual (Likes)
axes[0, 0].scatter(b_likes, pred_likes, alpha=0.4, color='royalblue', s=20)
axes[0, 0].plot([b_likes.min(), b_likes.max()],
                [b_likes.min(), b_likes.max()], 'r--', label='Perfect fit')
axes[0, 0].set_xlabel("Actual Likes")
axes[0, 0].set_ylabel("Predicted Likes")
axes[0, 0].set_title("Predicted vs. Actual (Likes)")
axes[0, 0].legend()

# Residual (Likes)
residuals_likes = b_likes - pred_likes
axes[0, 1].scatter(pred_likes, residuals_likes, alpha=0.4, color='royalblue', s=20)
axes[0, 1].axhline(0, color='red', linestyle='--')
axes[0, 1].set_xlabel("Predicted Likes")
axes[0, 1].set_ylabel("Residual")
axes[0, 1].set_title("Residual Plot (Likes)")

# Coefficients (Likes)
feature_names = ['Text Length', 'Hour of Day', 'Hashtag']
axes[0, 2].bar(feature_names, x_likes[1:], color='royalblue')
axes[0, 2].axhline(0, color='black', linewidth=0.8)
axes[0, 2].set_title("Coefficients (Likes)")
axes[0, 2].set_ylabel("Coefficient value")

# ── RETWEETS ─────────────────────────────────────────────────────────────────

# Predicted vs. Actual (Retweets)
axes[1, 0].scatter(b_retweets, pred_retweets, alpha=0.4, color='seagreen', s=20)
axes[1, 0].plot([b_retweets.min(), b_retweets.max()],
                [b_retweets.min(), b_retweets.max()], 'r--', label='Perfect fit')
axes[1, 0].set_xlabel("Actual Retweets")
axes[1, 0].set_ylabel("Predicted Retweets")
axes[1, 0].set_title("Predicted vs. Actual (Retweets)")
axes[1, 0].legend()

# Residual (Retweets)
residuals_retweets = b_retweets - pred_retweets
axes[1, 1].scatter(pred_retweets, residuals_retweets, alpha=0.4, color='seagreen', s=20)
axes[1, 1].axhline(0, color='red', linestyle='--')
axes[1, 1].set_xlabel("Predicted Retweets")
axes[1, 1].set_ylabel("Residual")
axes[1, 1].set_title("Residual Plot (Retweets)")

# Coefficients (Retweets)
axes[1, 2].bar(feature_names, x_retweets[1:], color='seagreen')
axes[1, 2].axhline(0, color='black', linewidth=0.8)
axes[1, 2].set_title("Coefficients (Retweets)")
axes[1, 2].set_ylabel("Coefficient value")

plt.tight_layout()
plt.savefig("trump_analysis.png", dpi=150)
plt.show()
```

Okay, that did not result in anything useful - except that there's seemingly no coolation, so I'll look into the data and find certain keywords to define (eg. "FAKE NEWS", "China") and make another design matrix with that in mind.

I will also transform the data into logarithmic data to make it scale better 

```python
# Defined additional keywords to include in the new model
keywords = [
    "FAKE NEWS",
    "Democrats",
    "China",
    "Hoax",
    "Military",
    "Border",
    "Great",
    "Corrupt",
    "Virus",
    "Vaccine",
     
]
b_likes = (np.array(Likes))
b_retweets =(np.array(Retweets))
# Boolean features for each keyword (case-insensitive)
keyword_features = {}
for kw in keywords:
    keyword_features[kw] = [1 if kw.lower() in t.lower() else 0 for t in text]

# the extended design matrix with the new keyword features
A_extended = np.column_stack([
    np.ones(len(text)),                          # bias
    text_length,                                 # tweet-længde
    hour_of_day,                                 # time på dagen
    hashtags,                                    # hashtag boolean
    *[keyword_features[kw] for kw in keywords]  # keyword booleans
])

# solve again with the new matrix
x_likes_ext,    _, _, _ = np.linalg.lstsq(A_extended, b_likes,    rcond=None)
x_retweets_ext, _, _, _ = np.linalg.lstsq(A_extended, b_retweets, rcond=None)

# Visualize the coefficients 
feature_names_ext = ['Text Length', 'Hour of Day', 'Hashtag'] + keywords

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Extended Model", fontsize=14)
# no need to show the bias term, so we skip the first coefficient which corresponds to the bias
axes[0].bar(feature_names_ext, x_likes_ext[1:], color='royalblue')
axes[0].axhline(0, color='black', linewidth=0.8)
axes[0].set_title("Likes")
axes[0].set_ylabel("Coefficient value")
axes[0].tick_params(axis='x', rotation=45)
# no need to show the bias term, so we skip the first coefficient which corresponds to the bias
axes[1].bar(feature_names_ext, x_retweets_ext[1:], color='seagreen')
axes[1].axhline(0, color='black', linewidth=0.8)
axes[1].set_title("Retweets")
axes[1].set_ylabel("Coefficient value")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig("trump_extended.png", dpi=150, bbox_inches='tight')

# Forudsagte værdier med den udvidede model
pred_likes_ext    = A_extended @ x_likes_ext
pred_retweets_ext = A_extended @ x_retweets_ext

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Trump Tweet engagement analysis - pt. 2", fontsize=14)

# Likes
axes[0].scatter(b_likes, pred_likes_ext, alpha=0.4, color='royalblue', s=20)
axes[0].plot([b_likes.min(), b_likes.max()],
             [b_likes.min(), b_likes.max()], 'r--', label='Perfect fit')
axes[0].set_xlabel("Actual Likes")
axes[0].set_ylabel("Predicted Likes")
axes[0].set_title("Predicted vs. Actual (Likes)")
axes[0].legend()

# Retweets
axes[1].scatter(b_retweets, pred_retweets_ext, alpha=0.4, color='seagreen', s=20)
axes[1].plot([b_retweets.min(), b_retweets.max()],
             [b_retweets.min(), b_retweets.max()], 'r--', label='Perfect fit')
axes[1].set_xlabel("Actual Retweets")
axes[1].set_ylabel("Predicted Retweets")
axes[1].set_title("Predicted vs. Actual (Retweets)")
axes[1].legend()

plt.tight_layout()
plt.savefig("trump_extended_pred.png", dpi=150, bbox_inches='tight')
plt.show()

# Print R² score for at måle hvor god modellen er
ss_res_likes = np.sum((b_likes - pred_likes_ext) ** 2)
ss_tot_likes = np.sum((b_likes - np.mean(b_likes)) ** 2)
r2_likes = 1 - ss_res_likes / ss_tot_likes

ss_res_rt = np.sum((b_retweets - pred_retweets_ext) ** 2)
ss_tot_rt = np.sum((b_retweets - np.mean(b_retweets)) ** 2)
r2_retweets = 1 - ss_res_rt / ss_tot_rt

print(f"R^2 Likes:    {r2_likes:.4f}")
print(f"R^2 Retweets: {r2_retweets:.4f}")
plt.show()
```

Based on this, I can conclude that at least in 2020, there's a corelation between:

- Certain words and likes/retweets — specifically "Virus" and "FAKE NEWS" show the largest positive coefficients for  likes, suggesting that emotionally charged language drives more engagement. This is a pattern that may partly explain the increasingly polarized political discourse on social media.

- It makes sense to transform data points for social media likes and retweets to a logarithmic model 

- Text length has almost no effect (the R^2 coefficent is close to zero in both models) 

- The model cannot reliably predict engagement. The Predicted vs. Actual plots show most points fall far below the perfect fit line, especially for high-engagement tweets. The model consistently underestimates viral tweets.

- Tweets with hashtags consistently got fewer likes and retweets, likely because hashtag-tweets tend to be formal endorsements rather than Trump's characteristically provocative posts that go viral

- The relationship is non-linear. A linear least squares model assumes a straight-line relationship between features and engagement, but in reality a single word like "Hoax" in a specific context may drive vastly different reactions, which a linear model cannot capture. 

In our HTX project, we computed the average likes per word using a dictionary, which resulted in a simple but unreliable result. A viral tweet inflated the average for any word used, which is what we also concluded back then. The Least Squared model fixes this by fitting of all features in a design matrix, allowing me to isolate the contribution of each feature while controlling for other. This is not possible with a simple word-average.
